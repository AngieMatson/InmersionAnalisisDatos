## **DESAFÍO CLASE 01**


1.   Promedio del área de todos los inmuebles en los barrios en el dataset. El top 10.
2.  Consultar otros datos estadísticos, conteo, mediana, valores mínimo y máximo.



```python
import matplotlib.pyplot as plt
```

```python
print(inmuebles.head())
```

```python
top_10_barrios = ['Usaquén', 'Zona Noroccidental', 'Bosa', 'Kennedy', 'Cedritos', 'Barrios Unidos', 'Engativa', 'Suba', 'Santa Barbara', 'Chapinero']

filtered_df = inmuebles[inmuebles['Barrio'].isin(top_10_barrios)]

promedio_area_por_barrio = filtered_df.groupby('Barrio')['Area'].mean()
```

```python
promedio_area_por_barrio = promedio_area_por_barrio.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
promedio_area_por_barrio.plot(kind='bar', color='purple')
plt.xlabel('Barrios del Top 10')
plt.ylabel('Promedio del Área')
plt.title('Promedio del Área de los Inmuebles en Barrios del Top 10')
plt.xticks(rotation=45)
plt.show()
```
